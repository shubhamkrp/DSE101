```
import pandas as pd
import numpy as np
import seaborn as sns

url="https://raw.githubusercontent.com/shubhamkrp/DSE101/main/pubg.csv"
pubg=pd.read_csv(url)
```

```
print(pubg.dtypes)
```

```
print(pubg.describe())
```

```
#this will give the average kills.
print(pubg['kills'].describe())
```

mean is 0.913400 => so an average persion kills 0.913400 many players, rounding off gives 1 kill for average player.

```
print(pubg['kills'].value_counts())
# 1% of total count i.e., 10000 is 100 which is done by 44+33+18+5
# total kills by 1% players is 7*44 + 8*33 + 9*18 + 11*5 = 799 kills.
total_kills = pubg['kills'].sum()
# So, 99% people have (total kills - 799) = (9134 - 799) = 8335
print(total_kills - 799)
```
